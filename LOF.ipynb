```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y = make_moons(n_samples=250,noise=0.05)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(X)
```

```python
dbcan = LocalOutlierFactor(n_neighbors=35)
```

```python
y_pred = dbcan.fit(x_scaled)
y_pred
```

```python
import numpy as np
outlier_indices = y_pred == -1
plt.scatter(X[:, 0], X[:, 1], color='k', label='Data points')
plt.scatter(X[outlier_indices, 0], X[outlier_indices, 1], color='r', s=50, label='Outliers')
```
